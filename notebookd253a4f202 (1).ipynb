```python
import numpy as np

i1, i2 = 0.05, 0.10

b1, b2 = 0.5, 0.7

np.random.seed(42)
W1 = np.random.uniform(-0.5, 0.5, (2, 2))  
W2 = np.random.uniform(-0.5, 0.5, (2, 2))  

print("Weights W1 (input->hidden):\n", W1)
print("Weights W2 (hidden->output):\n", W2)

inputs = np.array([i1, i2])

net_h = inputs @ W1 + b1          
out_h = np.tanh(net_h)           

print("\nnet_h:", net_h)
print("out_h:", out_h)

net_o = out_h @ W2 + b2            
out_o = np.tanh(net_o)             

print("\nnet_o:", net_o)
print("out_o (Network Output):", out_o)

targets = np.array([0.01, 0.99])  
E = 0.5 * np.sum((targets - out_o) ** 2)
print("\nTotal Error:", E)
```
